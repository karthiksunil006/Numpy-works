API --- Application Programming Interface
==========================================


It is communication between two different applications

The request and response between the two applications are transfered in json format

json is a lightweight format to transfer data across different applications

Example Scenario : Imagine going to a restaurant and the chef doesnt know your language. So the customer and chef are the two applications. In order to order food the customer gives the order to the waiter who gives the order to the chef in a language he understands. So the waiter acts as the json format
