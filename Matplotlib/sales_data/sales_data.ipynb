```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("sales_data_clean.csv")
df=pd.DataFrame(data)
df
```

```python
from matplotlib import pyplot

x=df["month_number"]
y=df["total_profit"]

pyplot.plot(x,y)
pyplot.xlabel("Month_number")
pyplot.ylabel("Total_Profit")
pyplot.title("Month_Number V/S Total_Profit")

pyplot.show()

```

```python
product_sales=[df["facecream"].sum(),df["facewash"].sum(),df["toothpaste"].sum(),df["bathsoap"].sum(),df["shampoo"].sum(),df["moisturizer"].sum()]

items=["facecream","facewash","toothpaste","bathsoap","shampoo","moisturizer"]

explode=(0,0,0.1,0,0,0)
colours=["red","cyan","green","brown","yellow","blue"]

pyplot.pie(product_sales,labels=items,colors=colours,explode=explode,autopct='%1.1f%%',startangle=90,shadow=True)

pyplot.show()
```
