```python
from matplotlib import pyplot
```

```python
Height=[176,178,158,160,167]
Weight=[68,70,58,60,57]

pyplot.plot(Height,Weight)

pyplot.xlabel("Height in cm")
pyplot.ylabel("Weight in kg")
pyplot.title("Height and Weight")

pyplot.show()
```

```python
pyplot.scatter(Height,Weight)

pyplot.xlabel("Height in cm")
pyplot.ylabel("Weight in kg")
pyplot.title("Height and Weight")

pyplot.show()
```

```python
velocity=[100,80,78,90,99]
displacement=[120,180,200,170,160]

pyplot.scatter(velocity,displacement,color="red")

pyplot.xlabel("Velocity in m/s")
pyplot.ylabel("Displacement in m")
pyplot.title("Velocity and Displacement")

pyplot.show()
```

```python
velocity=[100,80,78,90,99]
displacement=[120,180,200,170,160]

pyplot.plot(velocity,displacement,color="black")

pyplot.xlabel("Velocity in m/s")
pyplot.ylabel("Displacement in m")
pyplot.title("Velocity and Displacement")

pyplot.show()
```

```python
x1=[2,4,6,8,10]
y1=[1,3,5,7,9]

x2=[10,20,30,40,50]
y2=[5,15,20,25,35]

pyplot.plot(x1,y1,color="red")
pyplot.plot(x2,y2,color="green")

pyplot.xlabel("X axis------->")
pyplot.ylabel("Y axis------->")

pyplot.show()
```

```python
# Bar chart
x=[5,7,8,9,10]
y=[3,7,4,8,3]

pyplot.bar(x,y,color="green")

pyplot.xlabel("X axis")
pyplot.ylabel("Y axis")

pyplot.show()
```

```python
subjects=["Maths","Physics","Science","History","Computer Science"]

marks=[78,65,90,98,85]

pyplot.bar(subjects,marks,color="red")

pyplot.xlabel("Subjects---->",fontsize=15,color="green")
pyplot.ylabel("Marks----->",fontsize=15,color="green")
pyplot.title("Marks and Subjects")

pyplot.show()
```

```python
# Histogram
#===========

values=[1,2,3,56,23,30,7,8,9,10,11,125,150,14,15,16,17,18,19,20]

pyplot.hist(values)

pyplot.show()
```

```python
x=[1,1,2,2,2,4,4,4,5,5,5,5,6,6,6,6]

pyplot.hist(x)
pyplot.show()
```

```python
# Pie Chart
#===========

activities=["sleeping","travel","work","other"]
hours=[7,4,8,5]

pyplot.pie(hours,labels=activities,)
pyplot.show()
```

```python
# customizing our pie chart
#===========================

activities=["sleeping","travel","work","other"]
hours=[7,4,8,5]

colours=["red","cyan","green","brown"]
explode=(0.1,0,0,0)

pyplot.pie(hours,labels=activities,colors=colours,explode=explode,autopct='%1.1f%%',startangle=90,shadow=True)
pyplot.show()
```
