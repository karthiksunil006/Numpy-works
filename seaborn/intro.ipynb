SEABORN
========

Python data visualization library which write on top of matplotlib

Seaborn is more styled whereas matplotlib looks outdated

```python
import seaborn
import pandas as pd
from  matplotlib import pyplot
```

```python
days=["sun","mon","tue","wed","thur","fri","Sat"]

sales=[120,150,160,200,220,170,140]

data=pd.DataFrame({"Days":days,"Sales":sales})

seaborn.lineplot(data=data,x="Days",y="Sales",linestyle="--",marker="*")


pyplot.show()
```

```python
days=["sun","mon","tue","wed","thur","fri","Sat"]

sales=[120,150,160,200,220,170,140]

data=pd.DataFrame({"Days":days,"Sales":sales})

seaborn.lineplot(data=data)

pyplot.xlabel("days")
pyplot.ylabel("sales")
pyplot.title("Days V/S Sales")

pyplot.show()
```

```python
Employee_data={
    "dept":["Sales","It","Marketing","Testing","Hr"],
    "salaries":[20000,30000,15000,25000,35000]
}

df=pd.DataFrame(Employee_data)

seaborn.boxplot(x="dept",y="salaries",data=df,linecolor="red")
pyplot.title("Boxplot of Salaries v/s Dept")
pyplot.show()
```

```python
subjects=["physics","chemistry","biology","maths","history"]

marks=[70,80,90,96,85]

df=pd.DataFrame({"subjects":subjects,"marks":marks})


seaborn.barplot(x="subjects",y="marks",data=df,palette="coolwarm",hue="marks")
pyplot.title("Barplot of SUbjects V/S Marks")
pyplot.show()
```

```python
ages=[22,23,21,10,15,22,23,30,35,34,46]

seaborn.histplot(data=ages)

pyplot.title("Histogram of Ages")
pyplot.show()
```
