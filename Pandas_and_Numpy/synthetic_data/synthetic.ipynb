```python
import pandas
import numpy
```

```python
data=pandas.read_csv("synthetic_dataset.csv")

df=pandas.DataFrame(data)
df
```

In this dataset we can choose Rating as our predicting value 
And Category,Price,Stock,Discount as the features value

We chose rating as our prediction value because 

```python
df.isna().sum()
```

```python
df=df.drop("Category",axis=1) # Remove the coloumn Category as category is not important in finding the rating of the product. 
                              # The rating could be calculated using discount,price,stock. And also category has more than 50% missing values
                              # So it is better to completely remove the coloumn
                              # axis=1 means coloumn and axis=0 means row
df
```

```python
df.isna().sum()
```

```python
df=df.dropna(subset=["Rating"]) # drop the rows which has missing values for rating coloumn. because rating is our target value(so it cannot be removed completely)
                                # and has almost 50% missing values(so it is not ideal to fill these values with anything as it may reduce accuracy of model).  
df
```

```python
df.isna().sum()
```

```python
df["Price"]=df["Price"].fillna(df["Price"].mean())
df.isna().sum()
```

```python
df["Discount"]=df["Discount"].fillna(df["Discount"].mean())
df.isna().sum()
```

```python
df["Stock"]=df["Stock"].fillna(df["Stock"].mode()[0])
df.isna().sum()
```

```python
df
```

```python
# To change the object type data in Stock coloumn to numerical values so that we can train the machine
# This is called ENCODING
#                ========

df["Stock"]=df["Stock"].map({"In Stock":0,"Out of Stock":1})
df
```

```python
# Now that all the misiing values have been filled and encoding is done
# The next step is to bring all the coloumn values to a similar range
# This step is called SCALING
#                     =======

# Scaling >>> Getting all numerical values in each coloumn and convert them into similar range
# 
# Scaling can be done in two ways : 
# 
#  1) standard scaler
#     ===============
# Equation:
#   Z=(X-mean)/standard deviation


#  2) min max scaler
#     ============== 
```
