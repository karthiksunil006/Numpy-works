```python
# fill the missing values

df["trestbps"].fillna(df["trestbps"].mean(),inplace=True)
```

```python
df["restecg"].fillna(df["restecg"].mean(),inplace=True)
```

```python
df["thalach"].fillna(df["thalach"].mean(),inplace=True)
```

```python
df["ca"].fillna(df["ca"].mean(),inplace=True)
```

```python
df["thal"].fillna(df["thal"].mean(),inplace=True)
```

While filling missing values keep in mind,

if the total number of missing values is 300,400,500 or any value greater then it is not possible to fill these missing values. Even if we fill the missing values with mean,mode, or anything and train the model with these values then the final models prediction would be wrong and least accurate.

In such cases it is better to delete that coloumn from the dataset

if there are only few missing values we can fill these missing values with fillna() method and fill these with mean,mode,median
