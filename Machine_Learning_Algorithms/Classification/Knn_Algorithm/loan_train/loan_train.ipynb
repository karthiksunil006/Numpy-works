```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("loan_train.csv")
df=pd.DataFrame(data)
df
```

Target : Status
Input Features : Gender,Married,Dependents,Education,Self_Employed,Applicant_Income,Coapplicant_Income,Loan_Amount,Term,Credit_History,Area

```python
df=df.drop(columns=["Gender","Education"],axis=1)
df
```

```python
df.isna().sum()
```

```python
df.dtypes
```

```python
df["Married"]=df["Married"].fillna(df["Married"].mode()[0])
df.isna().sum()
```

```python
df["Dependents"]=df["Dependents"].fillna(df["Dependents"].mode()[0])
df.isna().sum()
```

```python
df["Self_Employed"]=df["Self_Employed"].fillna(df["Self_Employed"].mode()[0])
df.isna().sum()
```

```python
df["Term"]=df["Term"].fillna(df["Term"].mean())
df.isna().sum()
```

```python
df["Credit_History"]=df["Credit_History"].fillna(df["Credit_History"].mean())
df.isna().sum()
```

```python
df
```

```python
df.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder

Le=LabelEncoder()

df["Married"]=Le.fit_transform(df["Married"])
df["Dependents"]=Le.fit_transform(df["Dependents"])
df["Self_Employed"]=Le.fit_transform(df["Self_Employed"])
df["Area"]=Le.fit_transform(df["Area"])
df["Status"]=Le.fit_transform(df["Status"])
df.dtypes
```

```python
X=df.iloc[:,:-1]
Y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3)
```

```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()

X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn=KNeighborsClassifier()

model=knn.fit(X_train,Y_train)

Y_pred=knn.predict(X_test)
Y_pred
```

```python
from sklearn.metrics import accuracy_score

print(accuracy_score(Y_pred,Y_test))
```
