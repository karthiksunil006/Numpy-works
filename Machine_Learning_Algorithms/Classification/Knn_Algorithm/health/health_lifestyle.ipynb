.loc
=======

loc is a label based selector that lets you select specific rows and coloumns from your dataframe
It takes two arguments: row selector and column selector
These selectors can be labels, lists of labels, slices, or boolean masks
It returns the subset of the DataFrame matching those rows and columns

syntax: df.loc[row_Selector,coloumn_selector]

example: df.loc[['b', 'd'], ['name', 'city']] 
          what it does is that it selects rows b,d from a dataframe and selects coloumns name and city from the dataframe



example : df.loc[df["age"]>40,["age","gender","bmi"]]
         this is condition based selection where the rows are selected based on the condition "df["age"]>40" and specific coloumns "["age","gender","bmi"]" are selected

```python
df=df.drop("id",axis=1)
df=df.drop("age",axis=1)
df=df.drop("gender",axis=1)
df
```

```python
X=df.iloc[:,:-1]
Y=df.iloc[:,-1]
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3)

```

```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()
scaler.fit(X_train)
X_train=scaler.transform(X_train)
X_test=scaler.transform(X_test) 

X_train
                             
```

```python
X_test
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
knn=knn.fit(X_train,Y_train)

y_pred=knn.predict(X_test)  

print(y_pred)
```

```python
from sklearn.metrics import accuracy_score

print(accuracy_score(y_pred,Y_test))
```
