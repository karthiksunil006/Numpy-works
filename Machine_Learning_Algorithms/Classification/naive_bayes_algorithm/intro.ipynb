Naive Bayes Algorithm
======================

Used for Classification
=======================

Where prediction is based on probability based calculation
============================================================


Probability: Chance for occuring an event
=========================================

probability(head) = 1/2 = 0.5

Types of Probabilitiy
=====================

Marginal

>> Chances for occuring a single event

Joint

>> Chances for occuring both events A and B

Conditional

>> Based on one occurance of an event we predict the next one (Naive Bayes follows)
>> Chances of occuring event A where event B has already occured
>> Based on the result of B we need to calculate the probability of occuring event A

prior probabilty
============== 
p(Yes)=2/5
p(No)=3/5


conditionals
==============

 p(Weather=Rainy | Play=Yes) = 2/2 = 1

 p(Weather=Rainy | Play=No) = 0/3 =0

 p(temp=Hot | Play=yes) = 0/2 = 0

 p(temp=Hot | Play=No) = 2/3 = 0.67


Joint Likelihood
=================

 p(Yes|features(Weather=rainy | temp=hot)) = 1 * 0  = 0

 p(No|features(Weather=Rainy | temp=hot)) = 0 * 0.67  = 0


Unormalized Posteriors = Likelihood * Priors
=============================================

p(Yes|features(Weather=rainy | temp=hot)) = 1 * 0 * 2/5 = 0

 p(No|features(Weather=Rainy | temp=hot)) = 0 * 0.67 * 3/5 = 0
