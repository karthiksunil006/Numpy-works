```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("hypothyroid.csv")
df=pd.DataFrame(data)
df
```

```python
df=df.replace("?",np.nan)
```

```python
df
```

```python
df=df.drop(columns=["age","sex"],axis=1)
```

```python
df.isna().sum()
```

```python
df=df.drop(columns=["TBG"],axis=1)
```

```python
df.dtypes
```

```python
df["TSH"]=df["TSH"].fillna(df["TSH"].mode()[0])
df["T3"]=df["T3"].fillna(df["T3"].mode()[0])
df["TT4"]=df["TT4"].fillna(df["TT4"].mode()[0])
df["T4U"]=df["T4U"].fillna(df["T4U"].mode()[0])
df["FTI"]=df["FTI"].fillna(df["FTI"].mode()[0])
```

```python
df.isna().sum()
```

```python
from sklearn.preprocessing import LabelEncoder

Le=LabelEncoder()

for i in list(df):

    df[i]=Le.fit_transform(df[i])
```

```python
df.dtypes
```

```python
X=df.iloc[:,:-1]
Y=df.iloc[:,-1]
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3)
```

```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()

X_train=scaler.fit_transform(X_train)

X_test=scaler.transform(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier

model=DecisionTreeClassifier(criterion="entropy")

model=model.fit(X_train,Y_train)

Y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

print(accuracy_score(Y_test,Y_pred))
```
