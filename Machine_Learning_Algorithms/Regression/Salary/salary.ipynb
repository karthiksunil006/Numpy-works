```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("Salary.csv")
df=pd.DataFrame(data)
df
```

```python
X=df.iloc[:,:-1]
Y=df.iloc[:,-1]
X
```

```python
Y
```

```python
from matplotlib import pyplot

pyplot.scatter(X,Y,label="data points")
pyplot.xlabel("Years Experience")
pyplot.ylabel("Salary")
pyplot.title("Experience v/s Salary")
pyplot.legend()
pyplot.show()
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression

lr=LinearRegression()

lr=lr.fit(X_train,Y_train)

Y_pred=lr.predict(X_test)
```

```python
Y_pred
```

```python
Y_test
```

```python
pyplot.scatter(X_train,Y_train,color="red")
pyplot.plot(X_test,Y_pred)
pyplot.show()
```

```python
# y = mx + c

print(lr.coef_)  # this is m
print(lr.intercept_)  # this is c
```
