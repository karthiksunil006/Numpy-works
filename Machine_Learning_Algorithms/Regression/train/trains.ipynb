```python
# COnvert order date and ship date to date time

df['Ship Date'] = pd.to_datetime(df['Ship Date'], format="%d/%m/%Y")
df["Order Date"]=pd.to_datetime(df["Order Date"],format="%d/%m/%Y")
```

```python
df.dtypes
```

```python
# check missing values in each columns

df.isna().sum()
```

```python
# Rename all duplicate rows and show the new shape

df=df.drop_duplicates()
df.shape
```

```python
# # REname these columns for readability
#  Row id >> row_id
# Order id > order_id

df.rename(columns={"Order ID":"order_id"},inplace=True)
df.rename(columns={"Row ID":"row_id"},inplace=True)

df.columns
```

```python
# Sort the dataset by sales

df=df.sort_values(by="Sales",ascending=False)
df
```

```python
# Filter all orders where sales > 500

df[df["Sales"]>500]
```

```python
# Get all orders shipped using "Second Class" mode

df[df["Ship Mode"]=="Second Class"]
```

```python
# Select all orders from state="California"

df[df["State"]=="California"]
```

```python
# Select all orders from consumer segment with sales above 200

df.loc[(df["Segment"]=="Consumer")&(df["Sales"]>200),:]
```

```python
# Display all orders where category is "Furniture" and region="West"

df[(df["Category"]=="Furniture")&(df["Region"]=="West")]
```

```python
# find all orders where ship date is after order date

df[df["Ship Date"]>df["Order Date"]]
```

```python
# Create a new column Order_year from order date

df["Order_year"]=df["Order Date"].apply(lambda a:a.year)
df
```

```python
# create a new column SHipping_time_days Ship date-order date

df["Shipping_time_days"]=df["Ship Date"]-df["Order Date"]
df["Shipping_time_days"].apply
df
```

```python
# create a new column Profit_Category

df["Profit_Category"]=df["Sales"].apply(lambda a:"High" if a>500 else "Medium" if a>100 else "Low")
```

```python
# create a column that has the first name of the customer

df["first_name"]=df["Customer Name"].apply(lambda a:a.split(" ")[0])
df
```

```python
# create a new column discount_amount (discount=10% of sales)

df["Discount_amount"]=df["Sales"].apply(lambda a:a*(10/100))
df
```

```python
# Convert all values in City column to lowercase

cities=[]

for i in list(df["City"]):

    cities.append(i)

lowercase_cities=[]

for i in cities:

    lowercase_cities.append(i.lower())

lowercase_cities

df["City"]=lowercase_cities
df
```

```python
# Replace spaces in product name with underscores

df.rename(columns={"Product Name":"Product_Name"},inplace=True)
df
df["Product_Name"]=df["Product_Name"].str.replace(" ","_")
df
```
